```python
from datasets import load_dataset
```

```python
train_dataset = load_dataset("glue", "mrpc", split ="train").shuffle(42)
validation_dataset = load_dataset("glue", "mrpc", split ="validation").shuffle(42)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
for i in range(8):
  train_dataset.shard(num_shards=8, index = i, contiguous=False) \
               .save_to_disk("/content/drive/MyDrive/Federated_learning/Train_split/{}/train.hf".format(i))

  validation_dataset.shard(num_shards=8, index = i, contiguous=False).save_to_disk("/content/drive/MyDrive/Federated_learning/Validation_split/{}/validation.hf".format(i))
```
